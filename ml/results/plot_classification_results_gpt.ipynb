```python
import pandas as pd
import matplotlib.pyplot as plt

def plot(df, path):
    overall_df = df[df['Class'] == 'overall']

    # Set up the plot
    fig, ax = plt.subplots()

    # Plot the metrics for each test dataset
    width = 0.2
    x = range(len(overall_df['Test Dataset']))

    ax.bar(x, overall_df['Accuracy'], width, label='Accuracy')
    ax.bar([p + width for p in x], overall_df['Precision'], width, label='Precision')
    ax.bar([p + width*2 for p in x], overall_df['Recall'], width, label='Recall')
    ax.bar([p + width*3 for p in x], overall_df['F1-Score'], width, label='F1-Score')

    # Add labels
    ax.set_xlabel('Test Dataset')
    ax.set_ylabel('Scores')
    ax.set_title('Overall Scores by Test Dataset')
    ax.set_xticks([p + width for p in x])
    ax.set_xticklabels(overall_df['Test Dataset'])
    ax.legend()
    plt.savefig(path)
    # Show plot
    plt.show()
```

```python
df = pd.read_csv("gpt_baseline_classification_results.csv")
plot(df, 'gpt_baseline_metrics_histogram.png')
```
